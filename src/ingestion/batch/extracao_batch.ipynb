Armazenar na Bronze - Importando pathlib e datetime

Utilizando o arquivo paths.py para criar um diretorio para armazenar na camada bronze de maneira particionada.\
Não faremos alteração na bronze, somente converteremos para parquet porque é um formato mais eficiente que CSV, sem tratar qualquer valor original.\
Adicionaremos duas colunas de controle (data de ingestão e origem).

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent.parent.parent
sys.path.append(str(PROJECT_ROOT))

from src.utils.paths import BRONZE_DIR, SILVER_DIR, GOLD_DIR

print(BRONZE_DIR)
```

```python
from datetime import date

def salvar_bronze(df, nome_entidade):
    # Não mexe no DataFrame original, trabalha numa cópia
    df = df.copy()

    hoje = date.today().isoformat()
    pasta_destino = BRONZE_DIR / nome_entidade / f"ingestion_date={hoje}"
    pasta_destino.mkdir(parents=True, exist_ok=True)

    df["_ingestion_date"] = hoje
    df["_source"] = "basedosdados_batch"

    caminho_arquivo = pasta_destino / f"{nome_entidade}.parquet"
    df.to_parquet(caminho_arquivo, index=False)

    print(f"[{nome_entidade}] salvo em {caminho_arquivo} ({len(df)} linhas)")
```

```python
salvar_bronze(df_uf, "uf")
salvar_bronze(df_meta_brasil, "meta_alfabetizacao_brasil")
salvar_bronze(df_meta_uf, "meta_alfabetizacao_uf")
salvar_bronze(df_meta_municipio, "meta_alfabetizacao_municipio")
salvar_bronze(df_municipio, "municipio")
salvar_bronze(df_alunos, "dados_alunos")
```
